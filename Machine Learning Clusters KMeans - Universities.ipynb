```python
#importing KMeans
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
```

```python
model
```

```python
model.fit(data)
```

```python
model.labels_
```

```python
#converting 1D to 2D
cluster=pd.DataFrame(model.labels_)
```

```python
#adding cluster to data
```

```python
data['tire']=cluster
```
